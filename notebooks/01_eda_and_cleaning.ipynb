```python
import pandas as pd
import numpy as np
import os

raw_data_path = os.path.join('..', 'data', 'raw', 'all_seasons.csv')

df = pd.read_csv(raw_data_path)

print("Dataset Shape:", df.shape)
df.head()
```

```python
# 1. Look for missing (null) values across columns
print("--- Missing Values Report ---")
print(df.isnull().sum())

# 2. Check for duplicate rows
print(f"\nTotal Duplicate Rows: {df.duplicated().sum()}")

# 3. View descriptive statistics to spot unusual outliers (like 0 minutes played or extreme stats)
df.describe()
```

```python
# 1. Fill missing college values with 'None/International'
df['college'] = df['college'].fillna('None/International')

# 2. Drop the 'Unnamed: 0' index artifact column if it's there
if 'Unnamed: 0' in df.columns:
    df = df.drop(columns=['Unnamed: 0'])

print("Remaining Missing Values:")
print(df.isnull().sum().sum())
```

```python
# Filter for players with more than 15 Games Played (gp)
df_filtered = df[df['gp'] > 15].copy()

print(f"Original rows: {len(df)}")
print(f"Filtered rows (gp > 15): {len(df_filtered)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select only the numeric performance columns for correlation analysis
numeric_cols = ['age', 'player_height', 'player_weight', 'gp', 'pts', 
                'reb', 'ast', 'net_rating', 'usg_pct', 'ts_pct']

corr_matrix = df_filtered[numeric_cols].corr()

# Set up the plotting window size
plt.figure(figsize=(10, 8))

# Create the heatmap visual
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Display the chart
plt.title('NBA Player Statistics Correlation Matrix')
plt.show()
```
